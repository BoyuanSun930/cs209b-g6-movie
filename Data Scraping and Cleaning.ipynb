#### 5. data scraping directly from the website

```python
import requests
import operator
from tqdm import tqdm
from bs4 import BeautifulSoup as bs
```

```python
# all genres interested that have enough data in imdb by ascending order of amounts in the database

titles_count = dict(action = 38183, adventure = 18979, animation = 5600, biography = 5384, comedy = 80650, 
                   crime = 26145, drama = 157971, family = 12389, fantasy = 11577, history = 6083, horror = 24039,
                   music = 5095, musical = 8859, mystery = 12987, romance = 38932, sci_fi = 11315, sport = 3164,
                   thriller = 36569, war = 7929, western = 8089)
genres_count = sorted(titles_count.items(), key=operator.itemgetter(1))
genres = [item[0] for item in genres_count]
```

```python
genres
```

```python
def get_titles_from_genre(genres, amount = 300, page_number_limit = 100):
    """ To create a genre-balanced dataset, we can get the titles in each genre by the given amount
    INPUTS
    ------
    genres : list of interested genres
    page_number : number of pages wanted, one page has 50 titles, so page_number = 5 will give 250 movies per genre
    
    OUTPUTS
    -------
    dictionaries : genre as key, list of titles and ids as entries corresponding to the genre
    """
    all_ids = set()
    base_url = 'http://www.imdb.com/search/title'
    genre_dict = dict()
    id_dict = dict()
    for genre in genres:
        print('current genre: {}'.format(genre))
        genre_dict[genre] = []
        id_dict[genre] = []
        count = 0
        for num in range(1, page_number_limit + 1):
            params = dict(title_type = 'feature', genres = genre, page = str(num))
            r = requests.get(base_url, params=params)
            soup = bs(r.text,'lxml')
            for movie_title in soup.findAll('h3')[:50]:
                movie_id = movie_title.a['href'].split('/')[2]
                if movie_id not in all_ids:
                    genre_dict[genre].append(movie_title.a.text)
                    id_dict[genre].append(movie_id)
                    all_ids.add(movie_id)
                    count += 1
            if count >= 300:
                break
    return genre_dict, id_dict
```

```python
# scrape the titles along with the ids
all_titles, all_ids = get_titles_from_genre(genres)
```

```python
# there could be issues with online data scraping so we should verify that the movies are actually belong to the
# genre with the following function
def verify_genre(ids, genres = ['sport', 'music', 'biography', 'animation', 'history', 'war', 'western', 'musical',
 'sci_fi', 'fantasy', 'family', 'mystery', 'adventure', 'horror', 'crime', 'thriller', 'action', 'romance',
 'comedy', 'drama']):
    """ this function verifies if all of the movies are assigned to the correct genre
    INPUTS
    ------
    ids : all ids collected from the imdb website with genre as its keys
    genres : list of genres to be verified
    OUTPUTS
    -------
    ids of the movie that has not been correctly assigned
    """
    for genre in genres:
        print('current genre : {}'.format(genre))
        for movie_id in tqdm(ids[genre]):
            flag = True
            cur_genre = ia.get_title_genres(movie_id)['genres']
            cur_genre_lowercase = set(map(str.lower, cur_genre))
            if genre not in cur_genre_lowercase:
                print('Wrong genre assigned. ID : {}'.format(movie_id))
                flag = False
        if flag:
            print('The movies in genre {} are correctly assigned'.format(genre))
            print('-----------------------------------------------------------')
```

```python
# verify if we have the movies in their genre by the function
verify_genre(all_ids, genres = ['fantasy', 'family', 'mystery', 'adventure', 'horror', 'crime', 'thriller', 
                                'action', 'romance','comedy', 'drama'])
```

#### scrape the relevant movie information using the API

```python
def scrape_movie_info(ids):
    """ This function scrape the movie information given the movie ids dictionary
    INPUTS
    ------
    ids: dict, movie ids
    OUTPUTS
    -------
    pd dataframe with fields:
    1. title : str
    2. imdb_id : str, movie id
    3. year : int
    4. duration : int, length of the movie in minutes
    5. all_genre : list of str, all genres assigned to the movie
    6. plot : str, plot
    7. rating : float, user ratings
    8. top_rank : int, top ranking
    9. bottom_rank : int, bottom ranking
    10. rating_count : int, rating count
    11. meta_score : int, critic score
    12. meta_review_count : int, review count
    13. meta_user_rating_count : int
    14. meta_user_score : float
    15. genre : genre assigned to the movie
    """
    genres = ids.keys()
    movies_info_list = []
    
    # list of key names in each movie info dictionary
    base_list = ['title', 'year', 'runningTimeInMinutes']
    ratings_list = ['rating', 'ratingCount', 'topRank', 'bottomRank']
    meta_list = ['metaScore', 'reviewCount', 'userRatingCount', 'userScore']
    
    for genre in genres:
        print('current genre : {}'.format(genre))
        for movie_id in tqdm(ids[genre]):
            
            # new movie's info dictionary
            movie_dict = {}
            
            # movie info dictionary
            movie_info = ia.get_title(movie_id)
            
            # three seperate dictionaries containing info about the movie
            movie_base = movie_info['base']
            movie_ratings = movie_info['ratings']
            movie_meta = movie_info['metacriticScore']
            
            # decompose the data and construct the dictionary
            for key in base_list:
                if key in movie_base:
                    movie_dict[key] = movie_base[key]
                else:
                    movie_dict[key] = None
            
            for key in ratings_list:
                if key in movie_ratings:
                    movie_dict[key] = movie_ratings[key]
                else:
                    movie_dict[key] = None
                    
            for key in meta_list:
                if key in movie_meta:
                    movie_dict[key] = movie_meta[key]
                else:
                    movie_dict[key] = None
            
            # all_genre entry
            genre_list = ia.get_title_genres(movie_id)['genres']
            movie_dict['all_genre'] = genre_list
            
            # title and imdb_id
            movie_dict['title'] = movie_base['title']
            movie_dict['imdb_id'] = movie_id
            
            # all the plot summary
            if 'plot' in movie_info:
                movie_dict['plot'] = movie_info['plot']
            else:
                movie_dict['plot'] = None
                
#             movie_dict = dict(title = movie_base['title'], imdb_id = movie_id, year = movie_base['year'], 
#                              duration = movie_base['runningTimeInMinutes'], all_genre = genre_list,
#                              plot = movie_info['plot'], rating = movie_ratings['rating'],
#                              rating_count = movie_ratings['ratingCount'], meta_score = movie_meta['metaScore'],
#                              meta_review_count = movie_meta['reviewCount'],
#                              meta_user_rating_count = movie_meta['userRatingCount'],
#                              meta_user_score = movie_meta['userScore'], genre = genre)
            movies_info_list.append(movie_dict)
    return movies_info_list
```
